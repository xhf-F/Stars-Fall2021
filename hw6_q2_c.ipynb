```python
%pylab inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
```

To estimate the $Main-Sequence \ Lifetime$, we can consider the "H" mass fraction in the model. We will compare several estimate ways and see the difference. For each method value, I have noted it on the figure.

Method 1): Find the "H" depleted completely point, directly.

Method 2): Based on the theory, it defines the point where $10\%$ "H" feuled for "He" as the main-sequence over time.

Method 3): Based on the "Method 2", the nulear reaction energy suporting the lunimosity can write a time computing formula:    $t_{nuc} \approx \frac{10\% \times M \times 6 \times 10^{18} }{L} \approx 10^{10} (\frac{M}{M_\odot})^{-2.9}$ 

For a 1$M_{\odot}$ star :

```python
history_1M = Table.read("history_1M.out", format='ascii', #header_start=3, data_start=4,
                            names=['age (years)', 'mass (M_sun)', 'rho_c (g/cm^3)', 'T_c (K)', 'T_eff (K)', 'L (L_sun)', 'X(H1) center', 'mu approx'])
history_1M
```

```python
plt.plot(history_1M['age (years)'], history_1M['X(H1) center'], label = "X")
xlabel(r'age (years)')
ylabel(r'X(H1) center')
plt.title(r'$1M_{sun}$  main-sequence lifetime', y=1.03)

plt.vlines(0.8e10, -0.05, 0.7, color = "0.5", ls =":")
plt.hlines(0.63, 0.0, 1.75e10, color = "0.3", ls =":")
plt.vlines(0.108e10, -0.05, 0.7, color = "0.3", ls =":")
plt.vlines(1e10, -0.05, 0.7, color = "0.3", ls =":")

plt.text(0.83e10, 0.27, r"1) ~ 8.0 x $10^9$ years")
plt.text(0.11e10, 0.05, r"2) ~ 1.08 x $10^9$ years")
plt.text(1.05e10, 0.5, r"3) ~ 1.0 x $10^{10}$ years")
plt.text(0.12e10, 0.64, r"10$\%$ H depletion")

plt.legend()

ax = plt.twinx()
#ax.plot(history_1M['age (years)'], history_1M['L (L_sun)'], '-g')
ax.plot(history_1M['age (years)'], history_1M['mass (M_sun)'], '-r', label = "mass")
ax.set_ylabel('mass ($M_{sun}$)')
#ax.legend()

```

Because we're also interested in the variation of mass and luminosity which can give us some senses of the main-sequence lifetime, I plot the mass with the mass fraction of "H".
