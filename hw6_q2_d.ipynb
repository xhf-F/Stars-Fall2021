```python
%pylab inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
```

Use "history" files to plot the H-R diagram with $T_{eff}$ versus $L$

and

Use both "history" files and "profile" files to draw the burning stage beginning points with the "time header". Or for the "H" burning starting point, we can directly find within the "X(H1) center" column where the value is halfway depleted.

1.For a 1$M_{\odot}$ star :

```python
history_1M = Table.read("history_1M.out", format='ascii', #header_start=3, data_start=4,
                            names=['age (years)', 'mass (g)', 'rho_c (g/cm^3)', 'T_c (K)', 'T_eff (K)', 'L (L_sun)', 'X(H1) center', 'mu approx'])
history_1M
```

```python
#xx = np.flip(history_1M['T_eff (K)'])
xx = np.log10(history_1M['T_eff (K)'])
yy = np.log10(history_1M['L (L_sun)'])
plt.plot(np.log10(history_1M['T_eff (K)']), np.log10(history_1M['L (L_sun)']))
plt.xlim((max(xx)+0.1), (min(xx)-0.1))
#xx=np.log10(profile_1M_He['rho (g/cm^3)'])
#plt.plot(xx, ((2/3)*xx+5.3), label = "non-relativistic Fermi degenerate line")
xlabel(r'$log_{10} \ T_{eff} \ (K)$')
ylabel(r'$log_{10} \ L \ (L_{sun})$')
plt.title(r'$1M_{sun} \ \ H-R$ diagram with each burning stage', y=1.03)
plt.text(xx[13658-2], yy[13658-2], r'$\oplus$')
plt.text(3.82, 0.1, "H")
plt.text(xx[3354-2], yy[3354-2], r"$\otimes$ He")
plt.savefig('hw6_q2_d.jpg')
```
