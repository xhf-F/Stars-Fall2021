```python
%pylab inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
```

A helium flash is a shortly thermal runaway nuclear fusion of helium of low mass stars. 

and 

To deal with the Fermi energy and degenerate line, I use the formula I direved from our homework#3 which is $log T$ ~ $\frac{2}{3} log \rho + 5.3$ for non-relativistic one.

```python
profile_1M_He = Table.read("profile_1M_Heburn.out", format='ascii', #header_start=3, data_start=4,
                            names=['r (R_sun)', 'm (M_sun)', 'rho (g/cm^3)', 'T (K)', 'P (dyn/cm^2)', 'mu', 'mu_e', 'L (L_sun)', 'X', 'Y', 'grad_a', 'grad_T', 'q_nuc (erg/g/s)'])
profile_1M_He
```

```python
plt.plot(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['T (K)']), label = "He burning")
xx=np.log10(profile_1M_He['rho (g/cm^3)'])
plt.plot(xx, ((2/3)*xx+5.3), label = "non-relativistic Fermi degenerate line")


#plt.twinx(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['L (L_sun)']))
#plt.plot(), label = "He burning")

xlabel(r'$log_{10}\rho \ (g/cm^3)$')
ylabel(r'$log_{10}T \ (K)$')
plt.legend()
plt.title(r'$1M_{sun}$ He burning profile on $log_{10}T$ vs. $log_{10}\rho$', y=1.03)

plt.hlines(8.0, -4.0, 6, color = "0.5", ls =":" )
plt.hlines(7.8, -6.0, 4, color = "0.3", ls =":" )
plt.vlines(3.7, -3, 7.9, color = "0.5", ls = ":")

plt.text(1.5, 2, r"$\rho$ ~$10^{3.7}$")
plt.text(-5.5, 8., r"$T$ ~$10^{8}$")
plt.text(-7.8, 7.5, r"$T$ ~$10^{7.8}$")

#ax = plt.twinx()
#ax1.plot(x, y1, 'g-')L
#ax.plot(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['L (L_sun)']), '-g')
#ax.set_ylabel('L')

```

For this 1$M_{\odot}$ star, it may experience a core He flash since the "He" burning line cross the "degenerate" line at a relatively low density (less than $\rho$ ~$10^6$) and then it goes up to arrive at a T ~$10^8$ K. Then the "He" will be ignited and after this He Flash, it may go down to a white dwarf finally. 

And I am also interested in the variation of luminosity, so I draw this with "He" burning in the same figure. And it seems making sense.

```python
plt.plot(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['T (K)']), label = "He burning")
xx=np.log10(profile_1M_He['rho (g/cm^3)'])
plt.plot(xx, ((2/3)*xx+5.3), label = "non-relativistic Fermi degenerate line")


#plt.twinx(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['L (L_sun)']))
#plt.plot(), label = "He burning")

xlabel(r'$log_{10}\rho \ (g/cm^3)$')
ylabel(r'$log_{10}T \ (K)$')
plt.legend()
plt.title(r'$1M_{sun}$ He burning profile on $log_{10}T$ vs. $log_{10}\rho$', y=1.03)
ax = plt.twinx()
#ax1.plot(x, y1, 'g-')
ax.plot(np.log10(profile_1M_He['rho (g/cm^3)']), np.log10(profile_1M_He['L (L_sun)']), '-g', label ="luminosity")
ax.legend()
ax.set_ylabel(r"L ($L_{sun}$)")



```
